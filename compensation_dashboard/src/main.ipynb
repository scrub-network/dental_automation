```python
import pandas as pd
import random
import faker

# Initialize Faker to generate random names and emails
fake = faker.Faker()

# Define possible values for each metric
states_cities = ['New York, NY', 'Los Angeles, CA', 'Chicago, IL', 'Houston, TX', 'Phoenix, AZ']
job_titles = ['General Dentist', 'Orthodontist', 'Pedodontist', 'Periodontist', 'Endodontist']
job_types = ['DSO', 'Private Practice']
education_levels = ['DDS', 'DMD', 'DDS with Specialization', 'DMD with Specialization']
employment_types = ['Full-time', 'Part-time', 'Contract', 'Temporary']
practice_sizes = ['Small', 'Medium', 'Large']
rural_urban = ['Rural', 'Urban']
sponsorship = ['Yes', 'No']
language_skills = [1,2,3,4,5]

# Number of entries to generate
num_entries = 1000

# Generate dummy data
data = []
for _ in range(num_entries):
    entry = {
        'Name': fake.name(),
        'Email': fake.email(),
        'Location': random.choice(states_cities),
        'Job Title': random.choice(job_titles),
        'Job Type': random.choice(job_types),
        'Education': random.choice(education_levels),
        'Employment Type': random.choice(employment_types),
        'Practice Size': random.choice(practice_sizes),
        'Area Type': random.choice(rural_urban),
        'Sponsorship': random.choice(sponsorship),
        'English Proficiency': random.choice(language_skills),
        'Compensation': round(random.uniform(80000, 150000), 2)  # Random compensation in USD
    }
    data.append(entry)

# Create a DataFrame
df = pd.DataFrame(data)
df.to_csv("../data/dummy_data.csv", index=False)
```

```python
import pandas as pd
import random
import faker

# Reinitialize Faker to generate random names and emails
fake = faker.Faker()

# Define possible values for each metric
states_cities = ['New York, NY', 'Los Angeles, CA', 'Chicago, IL', 'Houston, TX', 'Phoenix, AZ']
job_titles = ['General Dentist', 'Orthodontist', 'Pedodontist', 'Periodontist', 'Endodontist']
job_types = ['DSO', 'Private Practice']
education_levels = ['DDS', 'DMD', 'DDS with Specialization', 'DMD with Specialization']
employment_types = ['Full-time', 'Part-time', 'Contract', 'Temporary']
practice_sizes = ['Small', 'Medium', 'Large']
rural_urban = ['Rural', 'Urban']
need_sponsorship = [True, False]
english_proficiency_levels = ['Native', 'Fluent', 'Proficient', 'Basic']
percent_production = [10, 15, 20, 25, 30, 40, 50, None, None, None]  # Percentages for production-related compensation
details_production = ["The production details", None, None]
percent_collection = [10, 15, 20, 25, 30, 40, 50, None, None, None]  # Percentages for collection-related compensation
details_collection = ["The collection details", None, None]
signing_bonus = [None, 5000, 10000, 15000]  # Possible signing bonus amounts
relocation_bonus = [None, 2000, 4000, 6000]  # Possible relocation bonus amounts
daily_rate_details = ["Draw", "Guaranteed", "Draw - Term Less than 1 year", "Draw - Term More than 1 year"]

# Number of entries to generate
num_entries = 1000

# Generate updated dummy data
updated_data = []
for _ in range(num_entries):
    annual_salary = round(random.uniform(80000, 150000), 2)  # Random compensation in USD
    daily_rate = round(annual_salary / 272, 2)  # Approximate number of workdays in a year

    entry = {
        'Name': fake.name(),
        'Email': fake.email(),
        'Location': random.choice(states_cities),
        'Job Title': random.choice(job_titles),
        'Job Type': random.choice(job_types),
        'Education': random.choice(education_levels),
        'Years of Experience': random.randint(0, 30),
        'Employment Type': random.choice(employment_types),
        'Practice Size': random.choice(practice_sizes),
        'Area Type': random.choice(rural_urban),
        'Need Sponsorship': random.choice(need_sponsorship),
        'English Proficiency': random.choice(english_proficiency_levels),
        'Annual Salary': annual_salary,
        'Daily Rate': daily_rate,
        'Daily Rate Details': random.choice(daily_rate_details),
        '% Production': random.choice(percent_production),
        'Details on % Production': random.choice(details_production),
        '% Collection': random.choice(percent_collection),
        'Details on % Collection': random.choice(details_collection),
        'Signing Bonus': random.choice(signing_bonus),
        'Relocation Bonus': random.choice(relocation_bonus)
    }
    updated_data.append(entry)

# Create an updated DataFrame
updated_df = pd.DataFrame(updated_data)
updated_df.to_csv("../data/updated_dummy_data.csv", index=False)
updated_df
```
