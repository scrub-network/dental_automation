```python
from bs4 import BeautifulSoup

def extract_links(html_content):
    soup = BeautifulSoup(html_content, 'html.parser')
    
    # Find all anchor tags
    links = soup.find_all('a')
    
    # Extract href attributes
    hrefs = ['https://www.indeed.com' + link['href'] for link in links if link.has_attr('href')]
    
    return hrefs


content = result['content']
soup = BeautifulSoup(content, 'html.parser')
main_divs = soup.find_all('li', class_='css-5lfssm eu4oa1w0')

hrefs = [extract_links(str(div)) for div in main_divs if extract_links(str(div)) != []]
hrefs = [item for sublist in hrefs for item in sublist if '.com/rc/' in item or '.com/company/' in item]
hrefs
# col_names = ["job_id", "location", "state", "date_posted", "employer", "job_type", "days_of_week", "description", "post_link"]
```
